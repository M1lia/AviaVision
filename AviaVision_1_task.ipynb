```python
file_source = '/content/dataset/oil_pollution_dataset/obj_train_data/'
file_destination = '/content/Data_set/train/labels/'


%cd /content/dataset/oil_pollution_dataset/obj_train_data
files = os.listdir()
for i in range(len(files)):
  if (files[i].endswith('.PNG')):
    img_png = Image.open('/content/dataset/oil_pollution_dataset/obj_train_data/' + str(files[i]))
    img_png.save('/content/Data_set/train/images/' + str(files[i][:-3] + 'jpg'))
    os.remove('/content/dataset/oil_pollution_dataset/obj_train_data/' + str(files[i]))
  else:
    shutil.move(file_source + files[i], file_destination)


    print(files[i])
```
